```python
from Actuator import Actuator
from FiberHolder import FiberHolder
import time
import pyfirmata
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Address port to fiber holder board
address = "COM6"
fiber_board = pyfirmata.Arduino(address)
fiber_holder1 = FiberHolder(fiber_board, 7, rotation_factor=1.02, orientation="clockwise")
fiber_holder2 = FiberHolder(fiber_board, 2)

#Address port to motorized stages
address = "COM5"
board = pyfirmata.Arduino(address)
actuator1 = Actuator(board=board, in1=2, in2=3)
actuator2 = Actuator(board=board, in1=4, in2=5)
actuator3 = Actuator(board=board, in1=12, in2=13)
```

```python
angle = 0
fiber_holder1.rotation_speed="SLOW"
fiber_holder1.rotate(angle)
# fiber_holder2.rotate(angle)
```

```python
%matplotlib tk
x = np.linspace(0, 30, 31) # x values from -5 to 5
y = norm.pdf(x, 18, 3)  
plt.ion()
position = []
transmission = []
graph = plt.plot(0, 0)[0]
plt.xlim(0,30)
plt.pause(1)
for i in range(len(x)):
    actuator1.move_to(x[i])
    position.append(x[i])
    transmission.append(y[i])
    graph.remove()
    graph = plt.plot(position, transmission, color='blue')[0]
    plt.xlim(0, 30)
    plt.ylim(0, np.max(y))
    plt.pause(0.25)
y_max = np.max(y)
y_max_loc = np.where(y==y_max)
actuator1.move_to(x[y_max_loc][0])
plt.vlines(x= y_max_loc, ymin = 0, ymax = y_max, color='black', linestyle='dashed')

time.sleep(1)
graph.remove()

x_fine = np.linspace(x[y_max_loc]-1, x[y_max_loc]+1, 21)
y_fine = norm.pdf(x_fine, 18, 3)

position_fine = []
transmission_fine = []

graph_fine = plt.plot(0, 0)[0]

for i in range(len(x_fine)):
    actuator1.move_to(x_fine[i][0])
    position_fine.append(x_fine[i])
    transmission_fine.append(y_fine[i])
    graph_fine.remove()
    graph_fine = plt.plot(position_fine, transmission_fine, color='blue')[0]
    plt.xlim(x_fine[0], x_fine[-1])
    plt.ylim(0, .2)
    plt.pause(0.25)

y_max_fine = np.max(y_fine)
y_max_fine_loc = np.where(y==y_max_fine)
actuator1.move_to(x[y_max_fine_loc][0])
```

```python
distances = np.linspace(0, 30, 7)
fiber_holder1.rotate(45)
fiber_holder1.rotation_speed="FAST"
for distance in distances:
    actuator1.move_to(distance)
    fiber_holder1.rotate(0)
    time.sleep(.25)
    fiber_holder1.rotate(90)
    time.sleep(.25)
    fiber_holder1.rotate(45)
    time.sleep(.25)
```
